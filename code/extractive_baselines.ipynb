```python
import pandas as pd
from summa.summarizer import summarize
from lexrank import LexRank
from lexrank.mappings.stopwords import STOPWORDS
import os
```

```python
doc = []
for i in os.listdir():
    if i[-1]!='v':
        continue
    df = pd.read_csv(i)
    doc.append(df['Sentence'])
lxr = LexRank(doc, stopwords=STOPWORDS['en'])
```

### Getting the Brief

```python
k = 0
for i in range(len(df1['Label'])):
    if df1['Label'][i]=='material fact' and k<3:
        k+=1
        print(df1['Sentence'][i]) 
```

```python
string = ''
for i in range(len(df1['Label'])):
    if df1['Label'][i]=='material fact':
        string= string + df1['Sentence'][i] + ' '

summarize(string, ratio=0.05)
```

```python
string = []
for i in range(len(df1['Label'])):
    if df1['Label'][i]=='material fact':
        string.append(df1['Sentence'][i])

x = lxr.get_summary(string, summary_size=2, threshold=.1)
print(x)
outfile=open('out.txt', 'w')
for i in x:
    print(i)
    outfile.write(i)
    outfile.write('\n')
```

### Getting the Issues Framed

```python
from tqdm import tqdm
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
arr = []
for i in range(len(df1['Label'])):
    if df1['Label'][i]=='issues framed':
        arr.append(df1['Sentence'][i])

out = model.encode(arr)
lxr.get_summary(arr, summary_size=3, threshold=.1)
```

```python
string = ''
for a in arr:
    string = string+a+' '

summarize(string, ratio=0.12)
```

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
```

```python
corpus_embeddings = out /  np.linalg.norm(out, axis=1, keepdims=True)

# Perform kmean clustering
clustering_model = AgglomerativeClustering(n_clusters=3) #, affinity='cosine', linkage='average', distance_threshold=0.4)
clustering_model.fit(corpus_embeddings)
cluster_assignment = clustering_model.labels_

clustered_sentences = {}
for sentence_id, cluster_id in enumerate(cluster_assignment):
    if cluster_id not in clustered_sentences:
        clustered_sentences[cluster_id] = []

    clustered_sentences[cluster_id].append(arr[sentence_id])

for i, cluster in clustered_sentences.items():
    print("Cluster ", i+1)
    print(cluster[0])
    print("")
```
